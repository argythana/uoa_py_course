## 4. Select (Locate) based on index, or "labels", or "position".
Recommended [answer](https://stackoverflow.com/a/31593712)  
Please pay attention to the examples understand the difference between `.loc` and `.iloc`.   
**Index for rows is first, then index for columns. (Rows, Columns).**

### Slice rows

````python
# first row
df.iloc[0]
````

````python
df.loc[0]  # same result different method. Be careful!
````

````python
# second row
df.iloc[1]
````

````python
# iloc gets rows (and/or columns) at integer locations. Works on position.
# Works on index, works with numbers.
# iloc is exclusive.
df.iloc[0:2]  # exclusive
````

````python
# Locate by name but use double brackets => This is a dataframe.
df[["Q2"]]
````

````python
type(df[["Q2"]])  # two brackets => dataframe
````

````python
# A list of [column names]
df[["Q2", "Q5", "Q3"]]
````

## 5. Alternative syntax to slice rows and columns.   
You may use similar notation to lists and strings indexing, without `.loc` nor `iloc` .

````python
# value of a single "value" in a row.
df.Q2[8]
````

````python
# Basic indexing as we do it in lists and strings works similarly on df rows.
# EXCLUSIVE because it works with ordered indexing in numbers.

df[10:13]
````

````python
# Basic indexing as we do it in lists and strings works similarly on df columns.
# EXCLUSIVE because it works with ordered indexing in number.
df.Q2[10:13]
````

````python
# Basic indexing as we do it in lists and strings works similarly on df columns.
# Same result as above, different syntax.
# EXCLUSIVE because it works with ordered indexing in numbers.

df["Q2"][10:13]
````

````python
# Basic indexing as we do it in lists and strings works similarly on df columns and rows.
# Same result as above, different syntax.
df[["Q1", "Q2"]][10:13]
````

````python
# Basic indexing as we do it in lists and strings works similarly on df columns and rows.
# Same result as above, different syntax.
df[10:13][["Q1", "Q2"]]
````

## 6. Transpose a df.
Sometimes it is needed.    
Here I use it to explain `.loc` and `.iloc` and to demonstrate row indices that have worded labels instead of numbers.

````python
# Nothing but a view.
df.transpose().head(4)  # alternative syntax: df.T
````

````python
# df has not changed in memory.
df
````

````python
# Nothing but a view.
df.T
````

````python
# df has not changed in memory.
df
````

````python
# Store a new name in memory and keep df as it is.
transposed_df = df.T  # T is an alias for the transpose() function

transposed_df
````

````python
# this works, is inclusive
transposed_df.iloc[0:4]
````

````python
# This does work with names of rows.
transposed_df.loc["Q1": "Q2"]
````

````python
# using ranges of [names of rows, names of columns]
transposed_df.loc["Q1": "Q4", 1:6]
````

````python
transposed_df.index
````

````python
transposed_df.columns
````

````python
transposed_df.loc["Q1"]
````

````python
# this works using a range of [names of rows]
transposed_df.loc["Q1": "Q4"]
````

````python
# this works # using selected names of rows
transposed_df.loc[["Q1", "Q4", "Q6"]]
````

#### Revision example

````python
# This does not work. Do you understand why?
# No We don't!
# +5 points if you find it.

transposed_df.loc["Q1", "Q2"]
````

````python
# Solution: double brackets. There is no index named ("Q1", "Q2").
# With a single bracket it is read as a single row name "Q1", not as list of row names.
# Then, after the comma, pandas looks for a column with name "Q2" but there is no such column in the transposed df.

# Now it locates a [list of names] for the rows.
transposed_df.loc[["Q1", "Q2"]]
````

````python
# locate a name for the row "Q1" and then, after the comma, locate the name of the column 2.
transposed_df.loc["Q1", 2]
````

#### Summary: a list of names for columns or rows needs the proper syntax.

````python
# Locates a [list of names] for the rows and a [list of names] for the columns.
transposed_df.loc[["Q1", "Q5"], [2, 3, 4, 5]]
````

````python
# Slice a range of names OF the rows and a range of names OF THE ROWS AGAIN.
# COLUMNS are not sliced with this syntax. Be very careful.
# SORRY, hope you don't hate me, after this example.

transposed_df[:"Q5"][3:6]
````
